```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://search.naver.com/search.naver?sm=tab_hty.top&where=news&query=한성기업&sort=1"
```

```python
original_html = requests.get(url)
```

```python
html = BeautifulSoup(original_html.text,"html.parser")
```

```python
html
```

```python
articles = html.select("div.group_news > ul.list_news > li div.news_area > a")
print(articles)
```

```python
news_title = []
for i in articles:
    news_title.append(i.attrs['title'])
print(len(news_title))
```

```python
news_url = []
for i in articles:
    news_url.append(i.attrs['href'])
print(news_url)
```

```python
contents = []
for i in news_url:
    news = requests.get(i)
    news_html = BeautifulSoup(news.text,"html.parser")
    contents.append(news_html.find_all("p"))
contents
```

```python
from selenium import webdriver
# mac Safari
driver = webdriver.Safari()
```

```python
driver.get(url)
```

```python
import time
a = driver.find_elements_by_css_selector('a.info')
naver_urls = []
# 위에서 생성한 css selector list 하나씩 클릭하여 본문 url얻기
count = 0
for i in a:
    count+=1
    i.click()

    # 현재탭에 접근
    driver.switch_to.window(driver.window_handles[count-1])
    time.sleep(3) #대기시간 변경 가능

    # 네이버 뉴스 url만 가져오기

    url2 = driver.current_url
    print(url2)

    if "news.naver.com" in url:
        naver_urls.append(url2)

    else:
        pass

    # 다시처음 탭으로 돌아가기(매우 중요!!!)
    driver.switch_to.window(driver.window_handles[0])
```
